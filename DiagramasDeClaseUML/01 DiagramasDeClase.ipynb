# 📘 Notas de Clase: Diagramas de Clases UML

## 1. Introducción

El **Diagrama de Clases** es uno de los diagramas estructurales más importantes de UML (Unified Modeling Language). Representa la estructura estática de un sistema mostrando sus clases, atributos, operaciones y relaciones.

---

## 2. Estructura de una Clase

Una clase en UML se representa como un rectángulo dividido en tres compartimentos:

| Compartimento | Contenido |
|---------------|-----------|
| **Superior** | Nombre de la clase |
| **Medio** | Atributos (propiedades) |
| **Inferior** | Operaciones (métodos) |

### Sintaxis de Atributos:
$$\text{visibilidad nombre : tipo [multiplicidad] = valor\_default \{propiedades\}}$$

### Sintaxis de Métodos:
$$\text{visibilidad nombre(parámetros) : tipo\_retorno \{propiedades\}}$$

---

## 3. Visibilidad

| Símbolo | Significado | UML | Código |
|---------|-------------|-----|--------|
| $+$ | Público | `+` | `public` |
| $-$ | Privado | `-` | `private` |
| $\#$ | Protegido | `#` | `protected` |
| $\sim$ | Paquete | `~` | `package` |

### 1. Herencia (Generalización)
La clase hija hereda atributos y métodos de la clase padre. Relación "es-un".



Símbolo: **<|--** (flecha vacía apuntando al padre)


### 2. Realización (Implementación)
Una clase implementa los métodos de una interfaz.


Símbolo: <|.. (flecha punteada vacía)


### 3. Composición
Relación fuerte **"parte-de"**. La parte no existe sin el todo. Si se destruye el contenedor, se destruyen las partes.


Símbolo: *-- (rombo negro)

### 4. Agregación
Relación débil **"tiene-un"**. Las partes pueden existir independientemente del todo.



Símbolo: **o--** (rombo blanco)


### 5. Asociación



Símbolo: **-->** (flecha simple) o **<-->** (bidireccional)


### 6. Dependencia

Relación de **uso temporal**. Una clase usa otra como parámetro, variable local o retorno.




Símbolo: **..>** (flecha punteada)

### Resumen

| Relación    | Símbolo | Fuerza     |
| ----------- | ------- | ---------- |
| Herencia    | `<\|--` | Fuerte     |
| Realización | `<\|..` | Fuerte     |
| Composición | `*--`   | Muy Fuerte |
| Agregación  | `o--`   | Débil      |
| Asociación  | `-->`   | Media      |
| Dependencia | `..>`   | Débil      |





### Jerarquía de Fuerza de Acoplamiento
- Dependencia<Asociacion<Agregacion<Composici<Herencia/Realizacion 

Buena práctica: Prefiere relaciones débiles (asociación, agregación) sobre fuertes (herencia, composición) para reducir el acoplamiento.
