```python
from IPython.display import HTML, display

dot_emoji = """
digraph CasosUso {
    rankdir=LR;
    splines=true;
    
    graph [nodesep=0.01, ranksep=0.5];
    
    node [fontname="Arial", fontsize=12];
    edge [arrowhead=none, color=black, minlen=1];
    
    Cliente [shape=none, label=<
        <TABLE BORDER="0" CELLBORDER="0" CELLSPACING="0">
            <TR><TD><FONT POINT-SIZE="60">游븴</FONT></TD></TR>
            <TR><TD>Cliente</TD></TR>
        </TABLE>
    >];
    
    UC1 [shape=ellipse, label="Realizar pedido", style=filled, fillcolor=lightblue1, width=2];
    UC2 [shape=ellipse, label="Cancelar orden", style=filled, fillcolor=lightblue1, width=2];
    
    Cliente -> UC1;
    Cliente -> UC2;
    
    labelloc=t;
    label="Sistema de Ventas";
    fontsize=14;
}
"""

display(HTML(f"""
<div style="text-align: center;">
    <!-- Bot칩n fuera del contenedor escalado -->
    <div style="margin-bottom: 20px; position: relative; z-index: 100;">
        <button onclick="guardarPNG()" style="padding: 10px 20px; background-color: #4CAF50; color: white; border: none; border-radius: 4px; cursor: pointer; font-size: 16px; position: relative; z-index: 101;">
            拘勇 Guardar PNG
        </button>
    </div>
    
    <!-- Gr치fico escalado separado -->
    <div style="transform: scale(1.5); display: inline-block; transform-origin: center top; margin-top: 20px;">
        <div id="graph"></div>
    </div>
</div>

<script src="https://unpkg.com/viz.js@2.1.2-pre.1/viz.js"></script>
<script src="https://unpkg.com/viz.js@2.1.2-pre.1/full.render.js"></script>
<script>
    var viz = new Viz();
    
    viz.renderImageElement(`{dot_emoji}`, {{format: 'png'}})
        .then(function(img) {{
            img.id = "diagrama-png";
            img.style.border = "1px solid #ddd";
            img.style.display = "block";
            document.getElementById('graph').appendChild(img);
        }});
    
    function guardarPNG() {{
        var img = document.getElementById("diagrama-png");
        if (img) {{
            var link = document.createElement("a");
            link.href = img.src;
            link.download = "diagrama_casos_uso.png";
            document.body.appendChild(link);
            link.click();
            document.body.removeChild(link);
        }} else {{
            alert("Espera a que cargue el diagrama...");
        }}
    }}
</script>
"""))
```
